# Comentario sobre el c칩digo

Este c칩digo realiza un an치lisis exploratorio de datos (EDA) para todas las hojas de un archivo Excel (`XLSX`). Utiliza las librer칤as **Pandas**, **Seaborn** y **Matplotlib** para analizar y visualizar datos. Su prop칩sito es extraer estad칤sticas b치sicas, identificar valores faltantes y duplicados, y detectar patrones o anomal칤as mediante gr치ficos. A continuaci칩n, se detalla el funcionamiento y los requisitos para ejecutarlo.

## Explicaci칩n del funcionamiento

1. **Importaci칩n de librer칤as**:
   - `pandas`: Para cargar los datos desde Excel y manipularlos.
   - `seaborn` y `matplotlib.pyplot`: Para realizar visualizaciones gr치ficas.

2. **Carga del archivo Excel**:
   - Obtiene los nombres de las hojas del archivo (`sheets`) usando `pd.ExcelFile(file_path).sheet_names`.

3. **Funci칩n `perform_eda`**:
   - Realiza el an치lisis exploratorio para un DataFrame espec칤fico.
     - **Informaci칩n b치sica del DataFrame**:
       - Usa `df.info()` para obtener el tipo de datos, columnas y n칰mero de registros.
     - **Valores faltantes**:
       - Analiza cu치ntos valores nulos (`NaN`) hay en cada columna con `df.isnull().sum()`.
     - **Registros duplicados**:

     - **Estad칤sticas descriptivas**:
       - Muestra estad칤sticas como promedio, desviaci칩n est치ndar, m칤nimos y m치ximos usando `df.describe()`.
     - **Visualizaci칩n de outliers**:
       - Crea gr치ficos de caja (boxplots) para las columnas num칠ricas.
     - **Distribuciones**:
       - Genera histogramas para mostrar la distribuci칩n de las variables num칠ricas.

4. **Iteraci칩n por hojas del archivo Excel**:
   - Para cada hoja, se carga en un DataFrame y se llama a la funci칩n `perform_eda` para ejecutar el an치lisis y las visualizaciones.

Adem치s, generar치 gr치ficos:

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar todas las hojas del archivo XLSX
file_path = "data/Internet.xlsx"
sheets = pd.ExcelFile(file_path).sheet_names  # Obtener nombres de las hojas

# Funci칩n de Exploratory Data Analysis (EDA)
def perform_eda(df, sheet_name):
    print(f"\n--- An치lisis de la hoja: {sheet_name} ---\n")
    
    # Informaci칩n general
    print("Informaci칩n del DataFrame:")
    print(df.info(), "\n")
    
    # Valores faltantes
    missing_values = df.isnull().sum()
    print("Valores faltantes por columna:\n", missing_values, "\n")
    
    # Valores duplicados
    duplicated_rows = df.duplicated().sum()
    print(f"Registros duplicados: {duplicated_rows}\n")
    
    # Estad칤sticas descriptivas
    print("Estad칤sticas descriptivas:\n", df.describe(), "\n")
    
    # Detecci칩n de outliers con boxplots
    numeric_cols = df.select_dtypes(include=['number']).columns
    for col in numeric_cols:
        plt.figure(figsize=(6, 4))
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot de {col} en {sheet_name}")
        plt.show()
    
    # Distribuci칩n de variables num칠ricas
    for col in numeric_cols:
        plt.figure(figsize=(6, 4))
        sns.histplot(df[col], bins=20, kde=True)
        plt.title(f"Distribuci칩n de {col} en {sheet_name}")
        plt.show()

# Iterar sobre cada hoja del archivo
for sheet in sheets:
    df = pd.read_excel(file_path, sheet_name=sheet)
    perform_eda(df, sheet)
````

# Volcado del contenido de data/Internet.xlsx

Este c칩digo integra un archivo Excel con m칰ltiples hojas (`sheets`) en una base de datos MySQL. Cada hoja del archivo es cargada como una tabla independiente en la base de datos. A continuaci칩n, se detalla c칩mo funciona el script y qu칠 requisitos son necesarios para ejecutarlo:

## Explicaci칩n del funcionamiento

1. **Librer칤as utilizadas**: 
   - `pandas`: Para cargar y manipular las hojas del archivo Excel.
   - `mysql.connector`: Para conectarse y operar con la base de datos MySQL.

2. **Cargar las hojas del archivo Excel**:
   - Se utiliza `pd.ExcelFile(file_path).sheet_names` para obtener los nombres de todas las hojas en el archivo Excel especificado.

3. **Conexi칩n a MySQL**:
   - Se establece una conexi칩n con la base de datos MySQL utilizando las credenciales proporcionadas (`host`, `user`, `password`, `database`).

4. **Procesamiento por hoja**:
   - **Renombrar columnas**: Los nombres de las columnas y de las tablas son limpiados para eliminar caracteres problem치ticos (`espacios`, `par칠ntesis`, `guiones`).
   - **Creaci칩n de tablas**: Se genera un comando SQL din치mico que crea una tabla para cada hoja, adaptando las columnas seg칰n los nombres del Excel. Todas las columnas son de tipo `VARCHAR(255)` para simplificar.
   - **Inserci칩n de datos**: Se recorre fila por fila para insertar los datos en las tablas correspondientes.

5. **Confirmar cambios**:
   - Los cambios son confirmados mediante `conexion.commit()`, y la conexi칩n y el cursor son cerrados al finalizar.

## Requisitos para ejecutarlo

### Prerrequisitos t칠cnicos:
- **Python instalado**: El script requiere Python y las librer칤as `pandas` y `mysql-connector-python`.
  - Instalar las librer칤as:
    ```bash
    pip install pandas mysql-connector-python
    ```
- **MySQL configurado**:
  - Base de datos MySQL accesible desde el sistema.
  - Configurar correctamente las credenciales (`host`, `port`, `user`, `password`) en el c칩digo.

### Archivo Excel:
- El archivo `Internet.xlsx` debe estar en la ruta especificada (`data/Internet.xlsx`).
- Debe tener datos estructurados y nombres de columnas que se puedan procesar.

### Permisos de MySQL:
- El usuario de la base de datos (`app-user-1`) debe tener permisos para:
  - Crear tablas (`CREATE`).
  - Insertar datos (`INSERT`).

## Advertencias:
- **Tipo de columnas**: Este script asigna `VARCHAR(255)` a todas las columnas, lo que puede no ser 칩ptimo para grandes vol칰menes de datos. Queda pendiente mejorar esto ajustando los tipos seg칰n el contenido con algun algoritmo que detecte el tipo de dato de toda la columna.
- **Riesgo de sobreescritura**: Si una tabla ya existe, los datos se insertar치n sin advertencia. Aseg칰rate de validar esto seg칰n las necesidades.

En resumen, este script automatiza la transferencia de datos desde Excel a MySQL, permitiendo la creaci칩n y carga din치mica de tablas basadas en las hojas del archivo.

````python
import pandas as pd
import mysql.connector

# Cargar todas las hojas del archivo XLSX
file_path = "data/Internet.xlsx"
sheets = pd.ExcelFile(file_path).sheet_names  # Obtener nombres de las hojas

# Conectar a MySQL
conexion = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="app-user-1",
    password="password123",
    database="Internet"
)
cursor = conexion.cursor()
for sheet in sheets:
    df = pd.read_excel(file_path, sheet_name=sheet)
    
    # Limpiar nombres de columnas para evitar errores en SQL
    df.columns = [col.replace(" ", "_").replace("(", "").replace(")", "").replace("-", "_") for col in df.columns]
    sheet = sheet.replace(" ", "_").replace("-", "_")  # Limpiar el nombre de la tabla

    # Crear la tabla en MySQL
    columnas_sql = ", ".join([f"`{col}` VARCHAR(255)" for col in df.columns])  # Usar `backticks` para proteger nombres
    sql_create_table = f"CREATE TABLE IF NOT EXISTS `{sheet}` ({columnas_sql})"
    
    print(f"Creando tabla: {sheet}")
    cursor.execute(sql_create_table)

    # Insertar los datos
    for _, row in df.iterrows():
        valores = tuple(row.fillna("").astype(str))  
        sql_insert = f"INSERT INTO `{sheet}` ({', '.join([f'`{col}`' for col in df.columns])}) VALUES ({', '.join(['%s']*len(row))})"
        cursor.execute(sql_insert, valores)

    print(f"Hoja '{sheet}' cargada exitosamente en MySQL.")

# Confirmar cambios y cerrar conexi칩n
conexion.commit()
cursor.close()
conexion.close()
````
